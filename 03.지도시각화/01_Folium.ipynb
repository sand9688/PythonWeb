```python
import folium
folium.__version__
```

```python
#open street map에 광주광역시청을 표시
map = folium.Map(location=[35.159844, 126.851784], zoom_start=12)
folium.Marker(
    location=[35.159844, 126.851784],
    popup=folium.Popup('광주역시 서구 내방로 111', max_width=200),
    tooltip='광주광역시청'
).add_to(map)
map
```

```python
map = folium.Map(location=[35.159844, 126.851784], zoom_start=13)
folium.Marker(
    location=[35.159844, 126.851784],
    popup=folium.Popup('광주광역시 서구 내방로 111', max_width=200),
    tooltip='광주광역시청'
).add_to(map)
folium.Marker(
    location=[35.146107, 126.92331],
    popup=folium.Popup('광주광역시 동구 서남로 1', max_width=200),
    tooltip='광주동구청'
).add_to(map)
folium.Marker(
    location=[35.152089, 126.88985],
    popup=folium.Popup('광주광역시 서구 농성동 경열로33', max_width=200),
    tooltip='광주서구청'
).add_to(map)
folium.Marker(
    location=[35.174100, 126.912157],
    popup=folium.Popup('광주광역시 북구 우치로 77', max_width=200),
    tooltip='광주북구청'
).add_to(map)
folium.Marker(
    location=[35.1329009, 126.902573],
    popup=folium.Popup('광주광역시 남구 봉선로 1', max_width=200),
    tooltip='광주남구청'
).add_to(map)
folium.Marker(
    location=[35.1395519, 126.7938068],
    popup=folium.Popup('광주광역시 광산구 광산로 29번길 15', max_width=200),
    tooltip='광주광산구청'
).add_to(map)
map
```

```python
map.save('광주시청 및 구청.html')
```
